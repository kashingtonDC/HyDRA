```python
# Compute the sums over hydrological year 05-06 and 16-17 (oct - oct)

sums0506 = []

for i in [prism0506, persia0506, trmm0506, chirps0506, dmet0506]:
    hydro_year = np.nansum(i[9:-3], axis = 0)
    hydro_year[hydro_year == 0] = np.nan
    sums0506.append(hydro_year)

sums1617 = []
    
for i in [prism1617, persia1617, trmm1617, chirps1617, dmet1617]:
    hydro_year = np.nansum(i[9:-3], axis = 0)
    hydro_year[hydro_year == 0] = np.nan
    sums1617.append(hydro_year)
    
diffs = []
for i,x in enumerate(sums0506):
    diffs.append(sums0506[i] - sums1617[i])
```

```python
def scale(X, x_min, x_max):
    nom = (X-np.nanmin(X, axis=0))*(x_max-x_min)
    denom = np.nanmax(X, axis=0) - np.nanmin(X, axis=0)
    return x_min + nom/denom 
```

```python
names = ['PRISM', 'PERSIANN', 'TRMM', 'CHIRPS', 'Daymet']
comparisons = [sums0506, sums1617, diffs, names]

nrows = len(comparisons[0])
ncols = len(comparisons)

vmin = 0

plt.figure(figsize=(10,5))

for idx,x in enumerate(comparisons[0]):
    sum1 = x
    sum2 = comparisons[1][idx]
    diff = comparisons[2][idx]
    scaled = scale(diff,-1,1)
    
    vmax = np.nanmax([sum1,sum2])
    
    plt.figure(figsize=(10,5))
    plt.suptitle(names[idx] + " P (mm)", size = 15)

    plt.subplot(1,3,1)
    plt.title("2005-2006")
    plt.imshow(sum1, vmin = vmin, vmax = vmax, cmap = "Blues")
    plt.colorbar(fraction=0.046, pad=0.04)
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.title("2016-2017")
    plt.imshow(sum2, vmin = vmin, vmax = vmax, cmap = "Blues")
    plt.colorbar(fraction=0.046, pad=0.04)
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.title("Pixelwise Difference (mm): 05/06 - 16/17")
    plt.imshow(diff, cmap = "jet")
    plt.colorbar(fraction=0.046, pad=0.04)
    plt.axis("off")

    plt.tight_layout()
    plt.show()
```

```python
# Compute the sums over hydrological year 05-06 and 16-17 (oct - oct)

swe_sums0506 = []

for i in [gldas_swe0506, fldas_swe0506, dmet_swe0506]:
    hydro_year = np.nansum(i[9:-3], axis = 0)
    hydro_year[hydro_year == 0] = np.nan
    swe_sums0506.append(hydro_year)

swe_sums1617 = []
    
for i in [gldas_swe1617, fldas_swe1617, dmet_swe1617]:
    hydro_year = np.nansum(i[9:-3], axis = 0)
    hydro_year[hydro_year == 0] = np.nan
    swe_sums1617.append(hydro_year)
    
swe_diffs = []
for i,x in enumerate(swe_sums1617):
    swe_diffs.append(swe_sums0506[i] - swe_sums1617[i])
```

```python
swe_names = ['GLDAS', 'FLDAS', 'Daymet']
swe_comparisons = [swe_sums0506, swe_sums1617, swe_diffs, swe_names]

nrows = len(comparisons[0])
ncols = len(comparisons)

vmin = 0

plt.figure(figsize=(10,5))

for idx,x in enumerate(swe_comparisons[0]):
    sum1 = x
    sum2 = swe_comparisons[1][idx]
    diff = swe_comparisons[2][idx]
        
    vmax = np.nanpercentile([sum1,sum2], 99)
    
    plt.figure(figsize=(10,5))
    plt.suptitle(swe_names[idx] + " SWE ($kg/m^2$)", size = 15)

    plt.subplot(1,3,1)
    plt.title("2005-2006")
    plt.imshow(sum1, vmin = vmin, vmax = vmax, cmap = "winter_r")
    plt.colorbar(fraction=0.046, pad=0.04)
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.title("2016-2017")
    plt.imshow(sum2, vmin = vmin, vmax = vmax, cmap = "winter_r")
    plt.colorbar(fraction=0.046, pad=0.04)
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.title("Pixelwise Difference (mm): 05/06 - 16/17")
    plt.imshow(diff, cmap = "cividis")
    plt.colorbar(fraction=0.046, pad=0.04)
    plt.axis("off")

    plt.tight_layout()
    plt.show()
```
