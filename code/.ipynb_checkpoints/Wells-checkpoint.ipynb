```python
import os
import numpy as np
import pandas as pd
import geopandas as gp
import matplotlib.pyplot as plt

from sklearn.gaussian_process import GaussianProcess
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel

from shapely.geometry import Point
```

```python
# Load the well data 

data_dir = "../data"
well_files = [os.path.join(data_dir,x) for x in os.listdir(data_dir) if "well" in x]

locs, data = sorted(well_files)
locs = pd.read_csv(locs)
data = pd.read_csv(data)

print( str(len(data['STN_ID'].unique())) + " unique wells")
print( str(len(locs['WELL_USE'].unique())) + " unique uses")
```

```python
# Turn into GP geometry and plot

lats = locs.LATITUDE
longs = locs.LONGITUDE
crs = {"init":"epsg:4326"}
geom = [Point(xy) for xy in zip(longs,lats)]
gdf = gp.GeoDataFrame(locs, crs = crs, geometry = geom)

gdf.plot()
plt.show()
```

```python
# Read cv shapefile
cv = gp.read_file("../shape/cv.shp")

# Extract wells in CV (spatial join)
within_cv = gp.sjoin(gdf, cv, how='inner', op='intersects')

within_cv.plot(figsize = (12,8), markersize = 0.1)
plt.show()
```

```python
# Join head data (well_measurements.csv) with wells in cv
cvdat = pd.merge(within_cv, data, left_on = "STN_ID", right_on = "STN_ID", how = "inner")

# Convert date col to datetime
cvdat.MSMT_DATE = pd.to_datetime(cvdat.MSMT_DATE)

# Sanity plot
plt.figure(figsize = (15, 8))
plt.plot(cvdat[cvdat.STN_ID == [cvdat.STN_ID.unique()[0]]].MSMT_DATE, cvdat[cvdat.STN_ID == [cvdat.STN_ID.unique()[0]]].RPE_WSE) 
```

```python
# Now, for each unique dates in casgem, extract the wells with data for that time step and perform 2d krigging

for i in cvdat.MSMT_DATE.unique()[:1]:
    selected = cvdat[cvdat.MSMT_DATE == i]
```
